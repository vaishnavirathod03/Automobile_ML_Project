```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('New_Car_Automobile.csv')
```

```python
data
```

```python
x=data.drop('price',axis=1)
```

```python
x
```

```python
y=data['price']
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
x[['car-name','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']]=x[['car-name','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']].apply(le.fit_transform)
```

```python
x
```

```python
y=le.fit_transform(y)
```

```python
y
```

```python
# Model Implementation --> Creation, Fit(Training) and Prediction
```

```python
# Importing the model
from sklearn.tree import DecisionTreeRegressor
```

```python
# Createing the object
dtc = DecisionTreeRegressor()     
```

```python
# Train the model using fit function
dtc.fit(x,y)
```

```python
# prredict on x dataset
y_pred = dtc.predict(x)
```

```python
y_pred
```

```python
y
```

```python
# Evaluate
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y,y_pred)
```
