```python
import pandas as pd
```

```python
data = pd.read_csv('New_Car_Automobile.csv')
```

```python
data.head(3)
```

```python
# loading the input variables in x excluding the class column as it is output
x=data.drop('car-name',axis=1)
```

```python
y=data[['car-name']]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
x[['fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']]=x[['fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']].apply(le.fit_transform)
```

```python
x
```

```python
y=le.fit_transform(y)
```

```python
y
```

```python
# Importing the model
from sklearn.tree import DecisionTreeClassifier
```

```python
# Creating the object
dtc = DecisionTreeClassifier(criterion='entropy')     #by default it takes gini
```

```python
# Train the model using fit function
dtc.fit(x,y)
```

```python
# prredict on x dataset
y_pred = dtc.predict(x)
```

```python
y_pred
```

```python
y
```

```python
from sklearn.metrics import classification_report,accuracy_score,recall_score,precision_score
from sklearn import metrics
```

```python
csr = classification_report(y,y_pred)
```

```python
print(csr)
```

```python
acc = accuracy_score(y,y_pred) 
```

```python
print("ACCURACY :",acc)
```

```python
rec=recall_score(y,y_pred,average='macro')
```

```python
print("RECALL_SCORE :",rec)
```

```python
prec=precision_score(y,y_pred,average='micro')
```

```python
print("PRECISION_SCORE :",prec)
```

```python
# Once metrics is imported we can use the confusion matrix function on our actual
# and predicted values.
confusion_matrix=metrics.confusion_matrix(y,y_pred)
```

```python
# To create a more interpretable visual display we need to convert the table 
# into a confusion matrix display.
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
```

```python
cm_display
```

```python
# Vizualizing the display requires that we import pyplot from matplotlib.
import matplotlib.pyplot as plt
```

```python
cm_display.plot()
plt.show()
```
