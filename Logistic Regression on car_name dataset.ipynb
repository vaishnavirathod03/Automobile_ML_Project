```python
import pandas as pd
```

```python
data=pd.read_csv('New_Car_Automobile.csv')
```

```python
data
```

```python
# loading the input variables in x excluding the class column as it is output
x=data.drop('car-name',axis=1)
```

```python
x
```

```python
y=data[['car-name']]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
x[['fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']]=x[['fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']].apply(le.fit_transform)
```

```python
x
```

```python
y=le.fit_transform(y)
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x,y)
```

```python
y_pred=lr.predict(x)
```

```python
y_pred
```

```python
x.shape
```

```python
lr.predict_proba(x)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,classification_report
```

```python
acc=accuracy_score(y,y_pred)
```

```python
acc
```

```python
rec=recall_score(y,y_pred,average='micro')
```

```python
rec
```

```python
prec=precision_score(y,y_pred,average='micro')
```

```python
prec
```
