```python
import pandas as pd
```

```python
data=pd.read_csv('New_Car_Automobile.csv')
```

```python
data
```

```python
# loading the input variables in x excluding the class column as it is output
x=data.drop('car-name',axis=1)
```

```python
x
```

```python
y=data[['car-name']]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
x[['fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']]=x[['fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']].apply(le.fit_transform)
```

```python
x
```

```python
y=le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
# Fitting Naive Bayes to the Training set  
from sklearn.naive_bayes import GaussianNB  
classifier = GaussianNB()  
classifier.fit(X_train, Y_train) 
```

```python
# Predicting the Test set results  
y_pred = classifier.predict(X_test) 
```

```python
y_pred
```

```python
#Making the Confusion Matrix  
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score
cm = confusion_matrix(Y_test, y_pred)
```

```python
cm
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
# To create a more interpretable visual display we need to convert the table 
# into a confusion matrix display.
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
```

```python
cm_display
```

```python
cm_display.plot()
plt.show()
```

```python
acc=accuracy_score(Y_test,y_pred)
```

```python
pre=precision_score(Y_test,y_pred,average='macro')
```
