```python
import pandas as pd
import numpy as np
```

```python
d1=pd.read_csv('New_Car_Automobile.csv')
```

```python
# low--> 0.0 to 0.3   (-0.3 to 0.3)
# moderate--> 0.3 to 0.7  (-0.7 to -0.3)
# high--> 0.7 to 1.0    (-1 to 0.7)

d1.corr()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
d1['output']=le.fit_transform(d1['car-name'])
```

```python
# input
x=d1.iloc[:,[7,8,9,11,14,16]]
```

```python
x
```

```python
# output
y=d1.iloc[:,[24]]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x[['curb-weight','engine-size']]=sc.fit_transform(x[['curb-weight','engine-size']])
```

```python
type(x)
```

```python
# creatig svc model
from sklearn.svm import SVC
```

```python
svc=SVC(random_state=0)
```

```python
# training
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

```python
y_test.values
```

```python
# performance
from sklearn.metrics import accuracy_score,recall_score,precision_score,classification_report,confusion_matrix
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
rec=recall_score(y_test,y_pred,average='macro')
```

```python
rec
```

```python
precision_score(y_test,y_pred,average='macro')
```

```python
c_report=classification_report(y_test,y_pred)
```

```python
c_report
```

```python
cm=confusion_matrix(y_test,y_pred)
```
